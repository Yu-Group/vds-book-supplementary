## Step 2: Loading the data

Since this data in its raw format is quite complex, loading a workable version of the data into our environment will involve loading and combining multiple different tables, and filtering to the relevant portion of the combined dataset.


### Combining the files into a single dataset


```python
# load in column names
column_names = pd.read_table("../data/features.txt", names=["i", "var"], delim_whitespace=True)

# there are some duplicated column names, so add an integer to differentiate the non-duplicated entries
column_names["duplicate_counter"] = column_names.groupby("var").cumcount() + 1
column_names["var_no_duplicates"] = column_names["var"]
# for the duplicated entries add an integer to differentiate them 
column_names.loc[column_names["duplicate_counter"] != 1, "var_no_duplicates"] = column_names["var_no_duplicates"] + "_" + column_names["duplicate_counter"].astype(str)
column_names
```

```python
# load in the training data and add column names
measurements_train_orig = pd.read_csv("../data/train/X_train.txt", 
                                      delim_whitespace=True, 
                                      names=column_names["var_no_duplicates"])
measurements_train_orig.head()
```

```python
# load the subject ID variable
subject_id = pd.read_csv("../data/train/subject_train.txt", delim_whitespace=True, names=["id"])
measurements_train = measurements_train_orig.copy()
measurements_train["id"] = subject_id["id"]
# rearrange so subject id is first
col_order = ["id"] + measurements_train.columns.drop("id").tolist()
measurements_train = measurements_train[col_order]
```
